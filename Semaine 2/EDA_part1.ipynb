 **Importation des librairies**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

## Chargement des datasets


```python
chemin_CICIDS = '../Data/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv'
chemin_UNSW   = '../Data/UNSW-NB15_1.csv'
chemin_LOGS   = '../Data/cybersecurity_threat_detection_logs.csv'


df_cicids = pd.read_csv(chemin_CICIDS)
df_unsw = pd.read_csv(chemin_UNSW)
df_logs = pd.read_csv(chemin_LOGS)
```

## CIC-IDS-2017 : FridayWorkingHoursAfternoonDDos 

```python
# Dimensions
print(f'Dimensions du dataset : {df_cicids.shape[0]: } lignes et {df_cicids.shape[1]} colonnes')
```

```python
print('Les 5 premières lignes :')
df_cicids.head()
```

```python
print('Types de données par colonne :')
print(df_cicids.dtypes.value_counts())
df_cicids.dtypes
```

Le dataset contient majoritairement des variables numériques.

Cela indique que la grande majorité des variables sont quantitatives, ce qui est cohérent avec des données de trafic réseau (comptage de paquets, durée, tailles, etc.)

```python
# Valeurs manquantes
nombre_manquants = df_cicids.isnull().sum().sum()
print(f'Nombre de valeurs manquantes : {nombre_manquants}')
```

```python
# Doublons
nombre_doublons = df_cicids.duplicated().sum()
print(f'Nombre de doublons : {nombre_doublons}')
```
